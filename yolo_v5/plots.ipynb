```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gtsdb_all_608 = pd.read_csv('yolov5/runs/train/gtsdb_all_608/results.csv')
gtsdb_all_416 = pd.read_csv('yolov5/runs/train/gtsdb_all_416/results.csv')
gtsdb_all_256 = pd.read_csv('yolov5/runs/train/gtsdb_all_256/results.csv')

gtsdb_well_represented_608 = pd.read_csv('yolov5/runs/train/gtsdb_well_represented_608/results.csv')
gtsdb_well_represented_416 = pd.read_csv('yolov5/runs/train/gtsdb_well_represented_416/results.csv')
gtsdb_well_represented_256 = pd.read_csv('yolov5/runs/train/gtsdb_well_represented_256/results.csv')

gtsdb_under_represented_608 = pd.read_csv('yolov5/runs/train/gtsdb_under_represented_608/results.csv')
gtsdb_under_represented_416 = pd.read_csv('yolov5/runs/train/gtsdb_under_represented_416/results.csv')
gtsdb_under_represented_256 = pd.read_csv('yolov5/runs/train/gtsdb_under_represented_256/results.csv')

gtsdb_all_608_more_img = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img/results.csv')
gtsdb_all_416_more_img = pd.read_csv('yolov5/runs/train/gtsdb_all_416_more_img/results.csv')
gtsdb_all_256_more_img = pd.read_csv('yolov5/runs/train/gtsdb_all_256_more_img/results.csv')

gtsdb_all_608_cfg_yolov5s = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5s/results.csv')
gtsdb_all_608_cfg_yolov5n = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5n/results.csv')
gtsdb_all_608_cfg_mv = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5_myversion/results.csv')
gtsdb_all_608_cfg_mv2 = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5_myversion2/results.csv')
gtsdb_all_608_cfg_mv3 = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5_myversion3/results.csv')
gtsdb_all_608_cfg_mv5 = pd.read_csv('yolov5/runs/train/gtsdb_all_608_more_img_cfg_yolov5_myversion5/results.csv')
```

Precision

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(gtsdb_all_608_more_img['               epoch'], 
        gtsdb_all_608_more_img['   metrics/precision'], 
        label = 'gtsdb_608', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_yolov5n['               epoch'], 
        gtsdb_all_608_cfg_yolov5n['   metrics/precision'], 
        label = 'gtsdb_608_5n', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_yolov5s['               epoch'], 
        gtsdb_all_608_cfg_yolov5s['   metrics/precision'], 
        label = 'gtsdb_608_5s', 
        linestyle = '-')


ax.plot(gtsdb_all_608_cfg_mv['               epoch'], 
        gtsdb_all_608_cfg_mv['   metrics/precision'], 
        label = 'gtsdb_608_cfg_mv1', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv2['               epoch'], 
        gtsdb_all_608_cfg_mv2['   metrics/precision'], 
        label = 'gtsdb_608_cfg_mv2', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv3['               epoch'], 
        gtsdb_all_608_cfg_mv3['   metrics/precision'], 
        label = 'gtsdb_608_cfg_mv3', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv5['               epoch'], 
        gtsdb_all_608_cfg_mv5['   metrics/precision'], 
        label = 'gtsdb_608_cfg_mv4', 
        linestyle = '-')


ax.set_xlabel('epoka')
ax.set_ylabel('precyzja')
ax.set_title('Wykres precyzji względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```

Recall

```python
fig, ax = plt.subplots(figsize = (10, 6))

# Plot precision for each dataset
ax.plot(gtsdb_all_608_more_img['               epoch'], 
        gtsdb_all_608_more_img['      metrics/recall'], 
        label = 'gtsdb_608', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_yolov5n['               epoch'], 
        gtsdb_all_608_cfg_yolov5n['      metrics/recall'], 
        label = 'gtsdb_608_5n', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_yolov5s['               epoch'], 
        gtsdb_all_608_cfg_yolov5s['      metrics/recall'], 
        label = 'gtsdb_608_5s', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv['               epoch'], 
        gtsdb_all_608_cfg_mv['      metrics/recall'], 
        label = 'gtsdb_608_cfg_mv1', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv2['               epoch'], 
        gtsdb_all_608_cfg_mv2['      metrics/recall'], 
        label = 'gtsdb_608_cfg_mv2', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv3['               epoch'], 
        gtsdb_all_608_cfg_mv3['      metrics/recall'], 
        label = 'gtsdb_608_cfg_mv3', 
        linestyle = '-')

ax.plot(gtsdb_all_608_cfg_mv5['               epoch'], 
        gtsdb_all_608_cfg_mv5['      metrics/recall'], 
        label = 'gtsdb_608_cfg_mv4', 
        linestyle = '-')


ax.set_xlabel('epoka')
ax.set_ylabel('czułość')
ax.set_title('Wykres czułości względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```

F1 score

```python
fig, ax = plt.subplots(figsize = (10, 6))

gtsdb_all_608_more_img['F1_score'] = 2 * gtsdb_all_608_more_img['   metrics/precision'] * gtsdb_all_608_more_img['      metrics/recall'] / (gtsdb_all_608_more_img['      metrics/recall'] + gtsdb_all_608_more_img['   metrics/precision'])
gtsdb_all_608_cfg_yolov5n['F1_score'] = 2 * gtsdb_all_608_cfg_yolov5n['   metrics/precision'] * gtsdb_all_608_cfg_yolov5n['      metrics/recall'] / (gtsdb_all_608_cfg_yolov5n['      metrics/recall'] + gtsdb_all_608_cfg_yolov5n['   metrics/precision'])
gtsdb_all_608_cfg_yolov5s['F1_score'] = 2 * gtsdb_all_608_cfg_yolov5s['   metrics/precision'] * gtsdb_all_608_cfg_yolov5s['      metrics/recall'] / (gtsdb_all_608_cfg_yolov5s['      metrics/recall'] + gtsdb_all_608_cfg_yolov5s['   metrics/precision'])
gtsdb_all_608_cfg_mv['F1_score'] = 2 * gtsdb_all_608_cfg_mv['   metrics/precision'] * gtsdb_all_608_cfg_mv['      metrics/recall'] / (gtsdb_all_608_cfg_mv['      metrics/recall'] + gtsdb_all_608_cfg_mv['   metrics/precision'])
gtsdb_all_608_cfg_mv2['F1_score'] = 2 * gtsdb_all_608_cfg_mv2['   metrics/precision'] * gtsdb_all_608_cfg_mv2['      metrics/recall'] / (gtsdb_all_608_cfg_mv2['      metrics/recall'] + gtsdb_all_608_cfg_mv2['   metrics/precision'])
gtsdb_all_608_cfg_mv3['F1_score'] = 2 * gtsdb_all_608_cfg_mv3['   metrics/precision'] * gtsdb_all_608_cfg_mv3['      metrics/recall'] / (gtsdb_all_608_cfg_mv3['      metrics/recall'] + gtsdb_all_608_cfg_mv3['   metrics/precision'])
gtsdb_all_608_cfg_mv5['F1_score'] = 2 * gtsdb_all_608_cfg_mv5['   metrics/precision'] * gtsdb_all_608_cfg_mv5['      metrics/recall'] / (gtsdb_all_608_cfg_mv5['      metrics/recall'] + gtsdb_all_608_cfg_mv5['   metrics/precision'])

ax.plot(gtsdb_all_608_more_img['               epoch'], gtsdb_all_608_more_img['F1_score'], label = 'gtsdb_608', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_yolov5n['               epoch'], gtsdb_all_608_cfg_yolov5n['F1_score'], label = 'gtsdb_608_5n', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_yolov5s['               epoch'], gtsdb_all_608_cfg_yolov5s['F1_score'], label = 'gtsdb_608_5s', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_mv['               epoch'], gtsdb_all_608_cfg_mv['F1_score'], label = 'gtsdb_608_mv1', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_mv2['               epoch'], gtsdb_all_608_cfg_mv2['F1_score'], label = 'gtsdb_608_mv2', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_mv3['               epoch'], gtsdb_all_608_cfg_mv3['F1_score'], label = 'gtsdb_608_mv3', linestyle = '-')
ax.plot(gtsdb_all_608_cfg_mv5['               epoch'], gtsdb_all_608_cfg_mv5['F1_score'], label = 'gtsdb_608_mv4', linestyle = '-')

ax.set_xlabel('epoka')
ax.set_ylabel('F1')
ax.set_title('Wartość F1 względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```
