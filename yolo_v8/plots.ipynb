```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gtsdb_all_608_more_img_yolov8n = pd.read_csv('runs/detect/gtsdb_all_608_more_img_yolov8n/results.csv')
gtsdb_all_608_more_img_yolov8s = pd.read_csv('runs/detect/gtsdb_all_608_more_img_yolov8s/results.csv')
```

```python
print(gtsdb_all_608_more_img_yolov8n.columns.tolist())
```

Precision

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(gtsdb_all_608_more_img_yolov8n['                  epoch'], 
        gtsdb_all_608_more_img_yolov8n['   metrics/precision(B)'], 
        label = 'gtsdb_608_yolov8n', 
        linestyle = '-')

ax.plot(gtsdb_all_608_more_img_yolov8s['                  epoch'], 
        gtsdb_all_608_more_img_yolov8s['   metrics/precision(B)'], 
        label = 'gtsdb_608_yolov8s', 
        linestyle = '-')

ax.set_xlabel('epoka')
ax.set_ylabel('precyzja')
ax.set_title('Wykres precyzji względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```

Recall

```python
fig, ax = plt.subplots(figsize = (10, 6))

ax.plot(gtsdb_all_608_more_img_yolov8n['                  epoch'], 
        gtsdb_all_608_more_img_yolov8n['      metrics/recall(B)'], 
        label = 'gtsdb_608_yolov8n', 
        linestyle = '-')

ax.plot(gtsdb_all_608_more_img_yolov8s['                  epoch'], 
        gtsdb_all_608_more_img_yolov8s['      metrics/recall(B)'], 
        label = 'gtsdb_608_yolov8s', 
        linestyle = '-')

ax.set_xlabel('epoka')
ax.set_ylabel('czułość')
ax.set_title('Wykres czułości względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```

F1 score

```python
fig, ax = plt.subplots(figsize = (10, 6))

gtsdb_all_608_more_img_yolov8n['F1_score'] = 2 * gtsdb_all_608_more_img_yolov8n['   metrics/precision(B)'] * gtsdb_all_608_more_img_yolov8n['      metrics/recall(B)'] / (gtsdb_all_608_more_img_yolov8n['      metrics/recall(B)'] + gtsdb_all_608_more_img_yolov8n['   metrics/precision(B)'])
gtsdb_all_608_more_img_yolov8s['F1_score'] = 2 * gtsdb_all_608_more_img_yolov8s['   metrics/precision(B)'] * gtsdb_all_608_more_img_yolov8s['      metrics/recall(B)'] / (gtsdb_all_608_more_img_yolov8s['      metrics/recall(B)'] + gtsdb_all_608_more_img_yolov8s['   metrics/precision(B)'])


ax.plot(gtsdb_all_608_more_img_yolov8n['                  epoch'], 
        gtsdb_all_608_more_img_yolov8n['F1_score'], 
        label = 'gtsdb_608_yolov8n', 
        linestyle = '-')

ax.plot(gtsdb_all_608_more_img_yolov8s['                  epoch'], 
        gtsdb_all_608_more_img_yolov8s['F1_score'], 
        label = 'gtsdb_608_yolov8s', 
        linestyle = '-')

ax.set_xlabel('epoka')
ax.set_ylabel('F1')
ax.set_title('Wartość F1 względem epok dla wyszkolonych modeli dla powiększonego zbioru danych')
ax.grid(True)
ax.legend()

ax.set_ylim(0, 1.0)
ax.set_yticks([i / 10 for i in range(11)])
ax.set_xlim(0, 200)

plt.show()
```
